```python
F.<x> = GF(2)[]
K.<a> = GF(2^163, modulus=x^163 + x^7 + x^6 + x^3 + 1)

# Generate random polynomials
poly1 = K.random_element()
poly2 = K.random_element()

# Multiply
product = poly1 * poly2

print("degree: {}".format(F(product).degree()))

print(f"Poly1: 0x{poly1.integer_representation():041x}")
print(f"Poly2: 0x{poly2.integer_representation():041x}")
print(f"Product: 0x{product.integer_representation():041x}")
```

```python
F.<x> = GF(2)[]
K.<a> = GF(2^31)

s1 = "74"
s2 = "06f9"

# Convert hex to field elements
f1 = K.fetch_int(int(s1.lower(), 16))
f2 = K.fetch_int(int(s2.lower(), 16))

print("Modulus: {} \n".format(K.modulus()))

# Perform various operations
f3 = f1 * f2
print("({}) * ({}) \n = {} \n = {:08x}".format(f1, f2, f3, f3.integer_representation()))
```
